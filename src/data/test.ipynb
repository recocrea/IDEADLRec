```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制球体
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, color='black', alpha=0.3)

# 绘制坐标轴
ax.quiver(0, 0, 0, 1, 0, 0, color='black', length=1.5)
ax.quiver(0, 0, 0, 0, 1, 0, color='black', length=1.5)
ax.quiver(0, 0, 0, 0, 0, 1, color='black', length=1.5)

# 绘制箭头
ax.quiver(0, 0, 0, 0, 0.5, 0.5, color='blue', length=0.7, arrow_length_ratio=0.5)
ax.quiver(0, 0, 0, 0.5, 0.5, 0, color='red', length=0.7, arrow_length_ratio=0.5)

ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])

plt.show()
```
